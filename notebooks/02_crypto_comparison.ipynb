```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
assets = ["BTC-USD", "ETH-USD"]
data = yf.download(assets, period="6mo")["Close"]
data.dropna(inplace=True)
data.head()
```

```python
data.plot(figsize=(12,5))
plt.title("BTC vs ETH - Precio de Cierre")
plt.ylabel("USD")
plt.grid(True)
plt.show()
```

```python
returns = data.pct_change().dropna()
returns.head()
```

```python
sns.heatmap(returns.corr(), annot=True, cmap="coolwarm")
plt.title("Correlacin BTC vs ETH")
plt.show()
```

```python
volatility = returns.rolling(14).std()

volatility.plot(figsize=(12,5))
plt.title("Volatilidad mvil (14 das)")
plt.grid(True)
plt.show()
```

## Conclusiones

- BTC y ETH muestran alta correlacin.
- ETH presenta mayor volatilidad en ciertos perodos.
- Ambos activos son adecuados para anlisis comparativo y gestin de riesgo.
