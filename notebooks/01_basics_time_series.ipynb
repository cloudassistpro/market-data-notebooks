```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("default")
```

```python
import yfinance as yf

btc = yf.download("BTC-USD", period="6mo", interval="1d")
btc.head()
```

```python
btc = btc.dropna()
btc.info()
```

```python
plt.figure(figsize=(12,5))
plt.plot(btc.index, btc["Close"])
plt.title("BTC-USD Precio de Cierre")
plt.xlabel("Fecha")
plt.ylabel("Precio (USD)")
plt.grid(True)
plt.show()
```

```python
btc["return"] = btc["Close"].pct_change()
btc.dropna(inplace=True)

btc[["Close", "return"]].head()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(btc["return"], bins=50, kde=True)
plt.title("Distribucin de Retornos Diarios BTC")
plt.show()
```

## Conclusiones

- BTC presenta alta volatilidad comparado con activos tradicionales.
- Los retornos diarios muestran colas pesadas.
- La volatilidad no es constante (heterocedasticidad).
