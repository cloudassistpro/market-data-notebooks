```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
eurusd = yf.download("EURUSD=X", period="6mo")
eurusd.dropna(inplace=True)
eurusd.head()
```

```python
plt.figure(figsize=(12,5))
plt.plot(eurusd.index, eurusd["Close"])
plt.title("EUR/USD Precio de Cierre")
plt.grid(True)
plt.show()
```

## Conclusiones

- EUR/USD presenta menor volatilidad que criptoactivos.
- Ideal para estrategias conservadoras.
- til para anlisis macroeconmico y gestin de riesgo.
